L'esercizio richiede di trasformare questo codice in un python script chiamato ```main.py``` che genera questo grafico e lo salva sul disco. 

Dopodiché, l'idea è quella di caricare questo codice in una nuova repository (per esempio ```my_first_repo```) che abbia questa struttura:

- datasets/OPSD_Germany_all.csv # contiene il dataset

Qui avete il risultato finale che dovreste ottenere: https://github.com/mttga/neural_academy_first_repo/tree/main

````python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv(
    'dataset/OPSD_Germany_all.csv',
    parse_dates=True,
    index_col='Date'
)
df['Consumption'].plot(figsize=(15, 6),
                       marker='.',
                       markersize=3,
                       linestyle='')
plt.title('Consumptions')
plt.show()
````
